# What the heck is Polynomial ? ðŸ˜°

<font color='blue'>*Some Calculus to jog our memory*</font>

The function 

\begin{align}
f(x) = A_0x_n + A_1x^{n-1}+\ldots+A_n,
\end{align}

where $n$ is an integer, is known as a polynomial; the number $n$ is called the *degree of the polynomial*.
Here, the coeffecients $A_0,A_1,\ldots,A_n$ are real or complex numbers.

<font color='bue'>Theorem 1 ( Remainder Theorem)</font> *Division of a polynomial $f(x)$ by $x-a$ yields a remainder equal to $f(a)$.*

**Proof :** The quotient obtained by the division of $f(x)$ by $x-a$ will be a polynomial $f_1(x)$ of degree one less than that of $f(x)$, and the remainder will be a constant $R$. We can thus write. 
\begin{align}
f(x) = (x-a)f_1(x) + R    --- (1)
\end{align}
This equality holds for all values of $x$ different from $a$.

Now let $x$ approach $a$. Then the limit on the left side of (1) willbe equal to $f(a)$, while the limit of the right side will be equal to $R$. Since the functions $f(x)$ and $(x-a)f_1(x)+R$ are equal for $x\neq  a$ their limits are likewise as $xâ†’a$, that is $f(a)=R$

<font color='blue'>Corollary</font>: *If $a$ is a root of the polynomial, that is, if f(a)=0 then $x-a$ divides $f(x)$ without remainder and hence, $f(x)$ represented in the form of a product.
\begin{align}
f(x) = (x-a)f_1(x),
\end{align}

where $f_1(x)$ is a polynomial.

Example: $f(x)=x^3-6x^2+11x-6$ for $x=1$
\begin{align}
&x^3-6x^2+11x-6=(x-1)(x^2-5x+6)
\end{align}

**Any number (real or complex) which, when substituted into the equaltion in place of $x$, converts the equaition into an identity is called a root of the equation.**

<font color='blue'>Theorem 2 ( Fundamental Theorem of Algebra)</font> *Every rationa integral function f(x) has at least one root,real or complex.*

<font color='blue'>Theorem 4</font>.*Every polynomial of degree $n$ may be factored into $n$ linear factors of the form $x-a$ and a factor equal to the coeffecient of $x^n$.*

**Proof** Let $f(x)$ be a polynomial of degree $n$.
\begin{align}
f(x) = A_0x_n + A_1x^{n-1}+\ldots+A_n,
\end{align}

By virtue of the fundamental theorem, this polynomial has at least one root; we denoate it by $a_1$. Then, by the corollary of the remainder thoerem, we can write. 
\begin{align}
f(x) = (x-a_1)f_1(x)
\end{align}

where $f_1(x)$ is a polynomial of degree $n-1$; $f_1(x) also has a root. 

We designate it be $a_2$. Then 
\begin{align}
f_1(x) = (x-a_2)f_2(x)
\end{align}

where $f_2(x)$ is a polynomial of degree $n-2$. Similarly,

\begin{align}
f_2(x) = (x-a_3)f_3(x)
\end{align}

Continuing this process of factoring out linear factors, we arrive at the relation 

\begin{align}
f_{n-1}(x) = (x-a_n)f_n(x)
\end{align}

where $f_n$ is a polynomial of degree zero, i.e some fixed number. 

This number is obviously equal to the coeffecient of $x^n$, that is, $f_n=A_0$.

On the basis of the equalities obtained we can write

\begin{align}
f(x) = A_0(x-a_1)(x-a_2)\ldots (x-a_n) -------- (2)
\end{align}

from the expression (2) it follows that the number $a_1,a_2, \ldots,a_n$ are roots of the polynomial $f(x)$, since upon the substitution $x=a_1,x=a_2,\ldots,x=a_n$ the right side, and hece the left becomes zero. 

Example: $f(x) = x^3-bx^2+11x-6$ becomes zero when $x=1,x=2,x=3$

Therefore, $x^3-bx^2+11x-6 = (x-1)(x-2)(x-3)$

No value $x=a$ that is different from $a_1,a_2,\ldots,a_n$ can be a root of the polynomial $f(x)$, since no factor on the right side of (2) vanishes when $x=a$

**A polynomial of degree n cannot have more than n distinct roots**.

# Interpolation. Lagrange's Interpolation Formula ðŸ˜±

Let it be established, in the study of some phenomenon, that there is a functional relationship between the quantities $y$ and $x$ which describes the quantitative aspect of the phenomenon; the function $y=Î¦(x)$ is **unknown**, but experiment has established the values of this function $y_0,y_1,y_2,\ldots,y_n$ for certain values of the argument $x_0,x_1,x_2,\ldots,x_n$, in the interval $[a,b]$.

The **problem** is to find a function which would represent the unknown function $y=Î¦(x)$ on the interval $[a,b]$ either exactly or approximately. In more abstract fashion the problem may be formulated as follows: 
"Given on the interval $[a,b]$ the values of an unknown function $y=Î¦(x)$ at $n+1$ distinct points $x_0,x_1,x_2,\ldots,x_n$.

\begin{align}
y_0 = \phi(x), y_1 = \phi(x_1),\ldots,y_n=\phi(x_n)
\end{align}

it is required to find a **polynomial $P(x)$** of degree $\leq n$ that approximates expresses the function $Î¦(x)$.

For such a polynomial, it is natureal to take a polynomial whose values at the points $x_0,x_1,x_2,\ldots,x_n$ **coincide** with the corresponding values $y_0,y_1,y_2,\ldots,y_n$ of the function $\phi(x)$. As shown the in diagram. 


Screen Shot 2022-01-25 at 7.09.48 PM.png

Then the problem, which is called the **problem of interpolating a function**, is formulated thus: for a given function $\phi(x)$ find a polynomial $P(x)$ of degree $\leq n$ which, for the given values of $x_0,x_1,x_2,\ldots,x_n$ will take on the values. 

\begin{align}
y_0 = \phi(x), y_1 = \phi(x_1),\ldots,y_n=\phi(x_n)
\end{align}

For the desired polynomial,take a polynomial of degree $n$ of the form 

\begin{align}
P(x) = &C_0(x-x_1)(x-x_2)\ldots(x-x_n)\\+&C_1(x-x_0)(x-x_2)\ldots(x-x_n)\\+
&C_2(x-x_0)(x-x_1)(x-x_3)\ldots(x-x_n) +\ldots\\+&C_n(x-x_1)(x-x_2)\ldots(x-x_{n-1}) ---- (1)
\end{align}

and define the coeffecients $C_0,C_1,\ldots,C_n$ so that the following condition are fulfilled. 

\begin{align}
P(x_0) = y_0, P(x_1)=y_1,\ldots,P(x_n) = y_n ------ (2)
\end{align}

(1)put $x=x_0$ then, taking into account equality (2), we get 

\begin{align}
y_0 = C_0(x_0-x_1)(x_0-x_2)\ldots(x_0-x_n)
\end{align}

whence
\begin{align}
C_0 = \frac{y_0}{(x_0-x_1)(x_0-x_2)\ldots(x_0-x_n)}
\end{align}

Then,setting $x=x1$, we get 
\begin{align}
y_1 = C_0(x_1-x_0)(x_1-x_2)\ldots(x_1-x_n)
\end{align}

whence
\begin{align}
&C_1 = \frac{y_1}{(x_1-x_0)(x_1-x_2)\ldots(x_1-x_n)}
\end{align}

In the same way we find 
\begin{align}
&C_2 = \frac{y_2}{(x_2-x_0)(x_2-x_1)\ldots(x_2-x_n)}\
&\ldots\\
&C_n = \frac{y_n}{(x_n-x_0)(x_n-x_2)\ldots(x_n-x_{n-1})}
\end{align}

Substituting these values of the coeffecients into (1), we get 

\begin{align}
P(x) &= \frac{(x-x_1)(x-x_2)\ldots(x-x_n)}{(x_0-x_1)(x_0-x_2)\ldots(x_0-x_n)}y_0 + \frac{(x-x_0)(x-x_2)\ldots(x-x_n)}{(x_1-x_0)(x_0-x_2)\ldots(x_0-x_n)}y_1 + \frac{(x-x_0)(x-x_1)\ldots(x-x_n)}{(x_2-x_0)(x_2-x_1)\ldots(x_0-x_n)}y_2 +\ldots + \frac{(x-x_0)(x-x_1)\ldots(x-x_{n-1})}{(x_n-x_1)(x_0-x_2)\ldots(x_n-x_{n-1})}y_n
\end{align}

This formula is called the **Lagrangian Interpolation Formula**

#Polynomial Regression ðŸ› 

**What is Polynomial regression ?**

Polynomial regression is a special case of linear regression where we fit a polynomial equation on  the data with a curvilinear relationship between the target variable and the independent variable. 

In a curvilinear relationship, the value of the target variable changes in a non-uniform manner w.r.t to the predictors(s).

Polynomial equation of degree $n$ represented as:    
\begin{align}
Y = Î¸_0 + Î¸_1 x +Î¸_2 x^2+ Î¸_3x^3+\ldots+Î¸_nx^n 
\end{align}

Here\
$Î¸_0$ : Bias\
$Î¸_1,Î¸_2,\ldots,Î¸_n$ : Weights  in the equation of the polynomial equation.\
$n$ : Degree of the polynomial

```python
class PolynomialRegression():

  def __init__(self,x,y,lr,iteration,degree):
    self.x = x
    self.y = y
    self.lr = lr
    self.iteration = iteration
    self.degree = degree

  def fit(self):
  
    self.loss_hist = []
    self.w = np.ones([self.degree+1,1])              #Initialize the weights
    self.y = self.y.reshape(self.y.shape[0],1)

    for i in range(0,self.iteration):
      self.y_pred = self.predict()
      loss = (1/self.x.shape[0])* 0.5 * np.sum((self.y_pred-self.y)**2)
      self.loss_hist.append(loss)
      error = self.y - self.y_pred
      self.w  += self.lr * np.dot(self.arr.T,error)* (1/self.x.shape[0])
    return self

  def X_transform(self):

    self.x = self.x.reshape(self.x.shape[0],1)
    self.arr = np.zeros([self.x.shape[0],self.degree+1])
    for i in range(0,self.degree+1):
      self.arr[:,i,np.newaxis] = self.x**i
    return self.arr

  def X_standardize(self):

    mean  = np.mean(self.arr[:,1:],axis=0)
    std = np.std(self.arr[:,1:],axis=0)
    self.arr[:,1:] = (self.arr[:,1:] - mean) / std
    return self.arr
  
  def predict(self):

    self.arr = self.X_transform()
    self.arr = self.X_standardize()
    #print(self.arr.shape)
    self.dot_prod = np.dot(self.arr,self.w)
    return self.dot_prod
```

## <font color='blue'>Interpolation and the Lagrance Polynomial</font>

One of the most useful and well-known classes of functions mapping the set of real numbers into itself, is the algenric polynomials, the set of functions of the form 

\begin{align}
P_n(x) = a_nx^n + a_{n-1}x^{x-1}+\ldots+a_1x +a_0
\end{align}

where $n$ is the non-negative integer and $a_0,\ldots,a_n$ are real constants. One reason for their importtance is that they uniformly approximate continuous functions. 

By this we mean that,given any function, defined and continuous on a closed and bounded interval, there exists a polynomial that is as "close" to the given function as desired. This result is expressed precisely in the Weierstrass Approximation Theorem.

**Weierstrass Approximation Theorem**

Suppose that $f$ is defined and continuous on $[a,b]$. for each $Ïµ>0$, there exists a polynomial $P(x)$, with the property that 

\begin{align}
|f(x) - P(x)| < Ïµ, \text{for all x in} [a,b]
\end{align}

Another important reason for considering the class of polynomials in the approaximation of functions is that the derivative and indefinite integeral of a polynomial are easy to determine and are also polynomials.

**Lagrance Interpolating Polynomials**

The problem of determining a polynomial of degree one that passes through the distinct points$(x_0,y_0)$ and $(x_1,y_1)$ is the same as approximating a function $f$ for which $f(x_0) = y_0$ and $f(x_1) = y_1$ by means of a first-degree polynomial **interpolating**, or agreeing with, the values of $f$ at the given points. 

Using polynomial for approximation within the interval given by the endpoints is called **polynomial interolation**.

Define the functions 
\begin{align}
L_0(x) = \frac{x-x_1}{x_0-x_1} \text{ and } L_1(x) = \frac{x-x_0}{x_1-x_0}
\end{align}
\
The linear **Lagrange interpolating polynomial** through $(x_0,y_0)$ and $(x_1,y_1)$ is
\
\
\begin{align}
P(x) &= L_0(x)f(x_0)+L_1(x)f(x_1)\\
&= \frac{x-x_1}{x_0-x_1}f(x_0) + \frac{x-x_0}{x_1-x_0} f(x_1)
\end{align}
\
$L_0(x_0)=1, L_0(x_1)=0, L_1(x_0)=0,L_1(x_1)=1$
\
\
Which implies that 

\begin{align}
P(x_0) &= 1.f(x_0)+0.f(x_1) = y_0\\
P(x_1) &= 0.f(x_0) +1.f(x_1) = y_1
\end{align}
\
So $P$ is the unique polynomial of degree at most one that passes through $(x_0,y_0)$ and (x_1,y_1)$

&L_0(x) = \frac{x-5}{2-5} = - \frac{1}{3} (x-5)\\
&L_1(x) = \frac{x-2}{5-2} = \frac{1}{3} (x-2)\\

so
